*While Loops*

---
*Comparing between for and While loops*

---
    For                              While

---



    for item in list_of_items:       While something_is_true:
        # Do something to each item     # Do something repeatedly

    for number in range(a,b):
        print(number)      

---

    While Loop

---
    While something_is_true:
        # Do this
        # Then do this
        # Then do this
        # until this condition becomes false at which it ends an exits the while loop.

---







```python
fruits = ["Apple", "Pear", "Orange"]
for fruit in fruits:
  print(fruit)
```

```python
for n in range(1, 6):
  print(n)
```


---
    Infinite Loop

---
    while 5 > 3:
        # Do this
        # Then do this
        # Then do this
        # five is always going to be larger than three until the end of time.
